# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
import matplotlib.pyplot as plt
# print(dir(pd))

df = pd.read_csv("summer.csv")

print(len(df["City"].unique()))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
# Design

GoldMedals = []

for i in df["Sport"].unique():
    
    Fm = pd.DataFrame(df[df["Sport"] == i])
    GoldMedals.append([len(Fm[Fm["Medal"] == "Gold"]), i])
    
GoldMedals.sort(reverse = True)
# print(GoldMedals)

Medals = [i[0] for i in GoldMedals[:5][:5]]
# print(Medals)
Medalists = [i[1] for i in GoldMedals[:5][:5]]
# print(Medalists)

# plot

plt.figure(figsize = (10, 5))
plt.bar(Medalists, Medals, color = "Green", width = 0.5)
plt.title("Gold Medalists")
plt.xlabel("Sports")
plt.ylabel("Medals")
plt.show()
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
# Design

Medals = []

for i in df["Sport"].unique():
    
    Fm = pd.DataFrame(df[df["Sport"] == i])
    Medals.append([len(Fm["Medal"]), i])
    
Medals.sort(reverse = True)
# print(Medals)

Medals = [i[0] for i in GoldMedals[:5][:5]]
# print(Medals)
Medalists = [i[1] for i in GoldMedals[:5][:5]]
# print(Medalists)

# plot

plt.figure(figsize = (10, 5))
plt.bar(Medalists, Medals, color = "Yellow", width = 0.5)
plt.title("Medalists")
plt.xlabel("Sports")
plt.ylabel("Medals")
plt.show()
```
